## Principais Fun√ß√µes e Conceitos

### üü¶ 1. `pd.DataFrame()`
- Cria uma tabela (linha e coluna).
- √â a estrutura principal do Pandas.

### üü¶ 2. `pd.read_csv()`
- L√™ arquivos `.csv` e transforma em DataFrame.

### üü¶ 3. `.head()`
- Mostra as primeiras linhas da tabela.

### üü¶ 4. `.tail()`
- Mostra as √∫ltimas linhas da tabela.

### üü¶ 5. `.info()`
- Mostra resumo das colunas, tipos e valores nulos.

### üü¶ 6. `.describe()`
- Mostra estat√≠sticas: m√©dia, m√≠nimo, m√°ximo etc.

### üü¶ 7. Sele√ß√£o de dados
- `df['coluna']` ‚Üí seleciona uma coluna.
- `df[['col1','col2']]` ‚Üí seleciona v√°rias colunas.
- `df.loc[linha]` ‚Üí busca por √≠ndice.
- `df.iloc[linha]` ‚Üí busca por posi√ß√£o.

### üü¶ 8. Filtros
- `df[df['idade'] > 18]`

### üü¶ 9. Ordena√ß√£o
- `df.sort_values('coluna')`

### üü¶ 10. Agrupamento
- `df.groupby('coluna').mean()`
